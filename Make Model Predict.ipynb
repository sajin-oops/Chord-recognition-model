```python
import librosa
import numpy as np
import joblib
```

```python
model = joblib.load("c_major_rf_model.joblib")
print("Model Loaded!")
```

```python
def extract_chroma(path):
    y,sr = librosa.load(path,sr=16000,mono = True)
    chroma = librosa.feature.chroma_stft(y=y,sr=sr)
    return np.mean(chroma,axis=1)
```

```python
# test_file = "C_3_EGC_S.wav"
test_file = r"C:\Users\AK Sajin\Sajin\Final year project\wav_dataset\C_major\C_3_EGC_S.wav"
feat = extract_chroma(test_file).reshape(1,-1)
pred = model.predict(feat)[0]
if pred == 1:
    print("prediction: C Major")
else:
    print("Prediction: Not C Major")
```
